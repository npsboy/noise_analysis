# Comprison of noises inside and outside city
In this project I am going to compare the max noises in the city and outside the city during the day, to find the best place to construct offices or work from home.

The data was reiceved from data.gov.in

```python
import matplotlib.pyplot as plt
import time
import pandas as pd
```

```python
def read_noise_data(noise_file): 
    
    noise_data = pd.read_csv(noise_file)
    
    #we are extracting only the date and maximum
    # sound in decibels during daytime.
    noise_data_max = noise_data.iloc[1:13,[0,4]]
    
    return noise_data_max
    
```

```python
noise_data_city = read_noise_data("city_noise.csv")
noise_data_outside_city = read_noise_data("outside_city_noise.csv")

noise_data_city.rename(columns={'Lmax': 'sound'}, inplace=True)
noise_data_outside_city.rename(columns={'Lmax': 'sound'}, inplace=True)
```

```python
def plot_noise_data(noise_data, plot_name):
    #noise_data.rename(columns={'Lmax': 'sound'}, inplace=True)
    noise_data.plot(x="Date",y="sound",kind='line')
    plt.axhline(y=80, color='r', label="Suitable for working")

    plt.title(plot_name)
    plt.xlabel("Date")
    plt.ylabel("Decibels")
    plt.legend()
    plt.show()   
```

```python
plot_noise_data(noise_data_city, "Noise levels inside city")
```

```python
plot_noise_data(noise_data_outside_city, "Noise levels outside city")
```

```python
def get_avg_noise(noise_data):
    avg = noise_data["sound"].mean()
    return avg
```

```python
city_avg = get_avg_noise(noise_data_city)
outside_city_avg = get_avg_noise(noise_data_outside_city)

df = pd.DataFrame({'location':['outside city', 'inside city'], 'sound':[outside_city_avg, city_avg]})
ax = df.plot.bar(x='location', y='sound', rot=0, color={"#032cfc","#fce303"})
plt.title("Average noise of city and outskirts of city")
plt.ylabel("sound in dB")
plt.show()
```

# Observation

- In both cases, city and outside city, the noise levels are almost always above 80 decibels which is the maximum noise level for working.

- February had the highest noise levels in both cases although the reason is unknown.

- The average sound produced in both cases are almost the same.

- The minimum levels in the outskirts of the city is lesser than the minimum levels in the city.

- Surprisingly, the maximum levels in the outskirts of the city is more than the maximum levels of the city.

# Conclusion
The outskirts of the city is a better place to work compared to the city.
